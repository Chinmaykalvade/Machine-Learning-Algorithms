```python
# Importing Libraries
import pandas as pd 
import numpy as np
```

```python
# Reading dataset and I have renamed the dataset to movie_data from exp_06_input
df=pd.read_csv("movie_data.csv")
```

```python
df.head()
```

```python
# Taking the name of movies to filter
names=df['movies'].values
```

```python
# Removing the number which is in the beginning of every movie to improve readability
l=[]
for name in names:
    for j in range (len(name)):
        if name[j]==":":
            l.append(name[j+1:])
            break
```

```python
# Stripping Unnecessary spaces
l=[name.strip(' ') for name in l]
len(l)
```

```python
# Replacing old column with cleaned values
df.movies=l
```

```python
df.head()
```

```python
# setting movies as index
df.set_index('movies',inplace=True)
```

```python
df.head()
```

```python
# setting name for columns
df.columns.names=['userId']
```

```python
# Taking transpose of dataset to find correlation
df=df.transpose()
```

```python
df.head()
```

```python
# filling NaN with zeros
df.fillna(0,inplace=True)
```

```python
# extracting ratings of a particular movie eg.Forest Gump
forrest_gump_ratings = df['Forrest Gump (1994)']
forrest_gump_ratings.head()
```

```python
# finding correlation of forest gump with other movies
movies_like_forest_gump = df.corrwith(forrest_gump_ratings)
movies_like_forest_gump.head()
```

```python
# making a dataframe of movies like forest Gump
corr_forrest_gump = pd.DataFrame(movies_like_forest_gump, columns=['Correlation'])  
corr_forrest_gump.dropna(inplace=True)  
corr_forrest_gump.head()
```

```python
# sorting the dataset to list most similar movies like forest Gump
# top 5 similar movies like Forest Gump
corr_forrest_gump.sort_values('Correlation', ascending=False).head(5)
```
