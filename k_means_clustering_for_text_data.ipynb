```python
# Category 0 is red
# Category 1 is green
# Category 2 is blue
colormap = {
    0: 'red',
    1: 'green',
    2: 'blue'
}

# Create a list of colors from every single row
colors = df.apply(lambda row: colormap[row.category], axis=1)

# And plot it!
ax = df.plot(kind='scatter', x='rafal', y='weaponri', alpha=0.1, s=300, c=colors)
ax.set_xlabel("rafal")
ax.set_ylabel("weaponri")
```
